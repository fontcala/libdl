# EXAMPLE  AutoEncoder_Overfit_CheckerboardPattern#
### In this example I just want to illustrate the use of "transposedConvLayer" to make an Autoencoder. The program uses various different layers, it just overfits an image with a few iterations and tries to replicate it in the output. Due to big "jump" in size at the maxpool and transposed convolution, it seems to generate a checkerboard pattern discussed here : https://distill.pub/2016/deconv-checkerboard/### Also can be seen here
https://towardsdatascience.com/review-deconvnet-unpooling-layer-semantic-segmentation-55cf8a6e380e
#### Network ####
Input(28,28,1) - Conv(5,5) - MaxPool (2,2) - Conv(3,3) - Out (12,12,5) - TransposedConv(3,3) - TransposedConv(2,2) - TransposedConv(5,5) - Output (28,28,1)
#### Note ####
This one is super hard to train, converges only a few times.  
#### Possible Improvements:  
Some smart learning rate decay strategy.

```python
# The library wrapper
import pybindings as pyb
  
```

```python
from matplotlib import pyplot as plt
%matplotlib inline 
from skimage import io
trXSub = np.load('sampleImage.npy') 
```

```python
AE = pyb.AutoEncoderExample(28,28,1,
                               12,36,
                               5,3,
                               0,0,
                               1,1)
```

```python
AE.Train(trXSub,trXSub,0.0000001,1000)
AEOut = AE.Test(trXSub)
```

```python
plt.title("test input")
plt.imshow(np.reshape(trXSub,(28,28)), interpolation='nearest', cmap='gray')
plt.show()
plt.title("test output")
plt.imshow(np.reshape(AEOut,(28,28)), interpolation='nearest', cmap='gray')
plt.show()
```
