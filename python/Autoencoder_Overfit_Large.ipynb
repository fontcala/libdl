# EXAMPLE  AutoEncoder_Overfit#
### In this example I just want to illustrate the use of "transposedConvLayer" to make an Autoencoder. The program uses 16 layers, it just overfits an image with a few iterations and tries to replicate it in the output. ###
#### Network ####
Input(28,28,1) - 8xConv(3,3) - Out (12,12,5) - 8xTransposedConv(3,3) - Output (28,28,1)
#### Note ####
In some rare occasions (less than 1/20) (probably depending on the initialization) the output turns out a completely black image local minima   
#### Possible Improvements:  
Some smart learning rate decay strategy.

```python
# The library wrapper
import pybindings as pyb
  
```

```python
from matplotlib import pyplot as plt
%matplotlib inline 
from skimage import io
trXSub = np.load('sampleImage.npy') 
```

```python
deepAE = pyb.AutoEncoderExample2(28,28,1,
                               3,3,3,3,3,3,3,3)
```

```python
deepAE.Train(trXSub,trXSub,0.00001,1000)
deepAEOut = deepAE.Test(trXSub)
```

```python
plt.title("test input")
plt.imshow(np.reshape(trXSub,(28,28)), interpolation='nearest', cmap='gray')
plt.show()
plt.title("test output")
plt.imshow(np.reshape(deepAEOut,(28,28)), interpolation='nearest', cmap='gray')
plt.show()
```
